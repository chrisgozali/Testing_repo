## 6. Calculate player "power".

---

Define a function that will calculate a player's "power". Player power is defined as the sum of the base statistics all of their pokemon.

$$
\text{player power } = \sum_{i = 1}^{n}\text{attack}_i + \text{defense}_i + \text{special attack}_i + \text{special defense}_i
$$

Where $i$ is an individual pokemon in a player's `player_pokemon`. ($\sum$ just means sum, so you're just adding up all the attributes listed above for all the pokemon in the player's `player_pokemon`).

Your function should:

*  Accept a `poke_players` dictionary and a `player_id` as arguments.
*  For the specified player_id, look up that player's pokemon.
*  Find and aggregate the attack and defense values for each of the player's pokemon.
*  Print "{player_name}'s power is {player_power}.", where the `player_power` is the sum of the base statistics for all of their pokemon.
*  Return the player's power value.

Check your work by looping through all players in your `poke_players` dict.

````python
poke_players[1]['player_name']
````

````python
#writing a function to calculate the sum of 4 different values in keys attack, defense, special attack and special defense for a specific player
def get_power(player_id, player_dict = poke_players):
    #set the total power to be 0 before entering the loop
    player_power = 0
    
    #loop through each of the player's pokemon and calculate the sum of their power
    for u in player_dict[player_id]['player_pokemon']:
        power = player_dict[player_id]['player_pokemon'][u]['attack'] + player_dict[player_id]['player_pokemon'][u]['defense'] + player_dict[player_id]['player_pokemon'][u]['special_attack'] + player_dict[player_id]['player_pokemon'][u]['special_defense']
        
        #add each pokemon's total power to the total player power
        player_power += power
    print (poke_players[player_id]['player_name'],'\'s power is ', player_power)
    return player_power
````

````python
#initialize dictionary representing desired attributes of pokemon
filter_options = {
    'Attack':   25,
    'Defense':  30,
    'Type':     'Electric'}
````

````python
#reference keys of filter options
list(filter_options.keys())
````

````python
def filtered_pokedex(pokedex_data,filter_options):
    #initialize an empty list of results
    filtered_results = []
    
    #need to write a nested for loop which iterates through attributes of each pokemon and compares them to the specified filter options
    #iterate through each pokemon's stats from pokedex
    for stats in range(1, len(pokedex_data.values())):
        
        #set a 0 counter to sum each criterion that is met from filter options. All criteria must be met for each pokemon to qualify
        count = 0
        
        #iterate through each filter option
        for keys, values in filter_options.items():
            
            #for the special case of pokemon type, where the value is a string (example: 'Electric')
            if isinstance(filter_options[keys],str):
                #check if the attribute matches with the pokemon in pokedex
                if filter_options[keys] == pokedex_data[stats][keys]:
                    count +=1
            #other attributes will be floats, hence a different criteria is set
            else:
                #check if the pokemon's stats meets the given criteria
                if filter_options[keys] <= pokedex_data[stats][keys]:
                    count +=1
        
        #only pokemon that meet all criteria from the filter will be added to the list of results 
        if count == len(filter_options):
            filtered_results.append(pokedex_data[stats])
            
    return filtered_results
````
